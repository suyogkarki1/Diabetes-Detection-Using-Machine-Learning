```python
import joblib
import pandas as pd

loaded_model = joblib.load('diabetes_non_diabetes_pipeline.pkl')
```

```python
custom_sample = pd.DataFrame([{
    'Age': 32,
    'Blood Glucose Reading': 82,
    'Diastolic Blood Pressure': 80,
    'Systolic Blood Pressure': 127,
    'Heart Rate': 90,
    'Body Temperature': 98.0,
    'SPO2': 97,
    'Gender': 'F',
    'Diastolic Blood Pressure Level': 'High',
    'Sweating  (Y/N)': 0,
    'Shivering (Y/N)': 0
}])

prediction = loaded_model.predict(custom_sample)[0]
probability = loaded_model.predict_proba(custom_sample)[0]

print("Prediction:", prediction)
print("Probability of Non-Diabetic:", probability[0])
print("Probability of Diabetic:", probability[1])

if prediction == 1:
    print("Final Prediction: Diabetic")
else:
    print("Final Prediction: Non-Diabetic")
```
